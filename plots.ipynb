```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/wvs/wave7.csv")
tf = pd.read_csv("./data/wvs/time_series.csv")
```

```python
COUNTRIES_ISO = ['AUS','CAN','DEU','NLD','USA']

corr_plot = ["national_distrust_index",
              "group_corruption",
              # migration
              "is_immigrant",
              "mother_immigrant",
              "father_immigrant",
              "migration_positive",
              "migration_negative",
              
              # political preference
              *[f"pol_value_{i}" for i in range(1,11) if i != 5],
              
              # hardships and security
              "hardships_questions",
              "security_neighborhood",
              "security_financial",
              "security_war",
              
              # demographics (including happiness)
              "age",
              "gender",
              "above_avg_inc",
              "happ__1",
              "happ__3",
              "happ__4",
              ]

df_corr = df.loc[:,corr_plot]
plt.figure(figsize=(12, 8)) 
sns.heatmap(df_corr.corr(), annot=False, cmap="coolwarm", linewidths=0.5, center = 0)
# Change the font of the tick labels (variables)
plt.xticks(fontsize=12) # Change font and size for x-axis labels (variables)
plt.yticks(fontsize=12)  # Change font and size for y-axis labels (variables)

plt.title('Heatmap of Feature Correlation')
plt.show()
plt.savefig("./plots/corr_map.png")
```

## Boxplots
### Visualizing different influences in the cross sectional data

```python
plt.figure(figsize=(8,4), dpi=300)
plt.rcParams["font.family"] = "sans-serif"
flierprops = dict(marker='x', markersize=5, markeredgecolor='gray')
sns.boxplot(data=df.loc[df.Q240 > 0], x="Q240", y="national_distrust_index", saturation=1 , color="skyblue", flierprops=flierprops, linewidth=.7,
            linecolor="black")
plt.xlabel("Political Orientation (1=Extreme Left, 10=Extreme Right)")
plt.ylabel("National Distrust Index")
plt.title("Box plot of Poltical Orientation by National Distrust")
plt.show()
plt.savefig("./plots/box_pol_or.png")
```

```python
plt.figure(figsize=(8,4), dpi=300)
plt.rcParams["font.family"] = "sans-serif"
flierprops = dict(marker='x', markersize=5, markeredgecolor='gray')
sns.boxplot(data=df.loc[df.Q288 > 0, :], x="Q288", y="national_distrust_index", saturation=1 , color="skyblue", flierprops=flierprops, linewidth=.7,
            linecolor="black")
plt.xlabel("Income Groups (1=Very low, 10=Very high)")
plt.ylabel("National Distrust Index")
plt.title("Box plot of income by national distrust")
plt.show()
plt.savefig("./plots/box_inc.png")
```

## time trend for distrust in Germany over the last waves

```python
plt.figure(figsize=(8,4), dpi=300)
plt.rcParams["font.family"] = "sans-serif"
flierprops = dict(marker='x', markersize=5, markeredgecolor='gray')
sns.boxplot(data=tf, x="wave", y="national_distrust_index", saturation=1 , color="skyblue", flierprops=flierprops, linewidth=.7,
            linecolor="black")
plt.xlabel("Wave")
plt.ylabel("National Distrust Index")
plt.title("Box plot of income by national distrust")
plt.show()
plt.savefig("./plots/box_wave.png")
```
